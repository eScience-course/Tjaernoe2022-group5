```python
from miditime.miditime import MIDITime
import numpy as np
import pandas as pd
```

```python
def to_midi(infile,nm):
    mymidi = MIDITime(160, nm+'.mid')
    tt = pd.read_csv(infile,skiprows=1,header=None,index_col=0)
    tt = np.array(tt)
    tt = tt.tolist()
    # Add a track with those notes                                                                                                                                                                              
    mymidi.add_track(tt)
    # Output the .mid file                                                                                                                                                                                      
    mymidi.save_midi()
```

```python
to_midi('arc_tas_music_585','arc_tas_music_585')
```
